#### Question 4: Linear classifiers

You have a data set with two input features x and y with a binary classification (T,F). You have the following data points: (x = 4, y= 2,F), (x=2,y=5,F), (x=6,y=4,T), (x=4,y=6,T). Which of the following lines separates the data best?

a) y = -2x-12

b) y=2x-12

c) y=2x+12

d) y = -2x+12

```python

x = [4,2,6]
y = [2,5,4]

plt.plot(x,y, "ro")
x = np.array([[4,2],[2,5],[6,4]])
y = - (2 * x) - 12
plt.plot(x,y)
plt.show()

x = np.array([[4,2],[2,5],[6,4]])
y =  (2 * x) + 12
plt.plot(x,y)
plt.show()

x = np.array([[4,2],[2,5],[6,4]])
y = - (2 * x) + 12
plt.plot(x,y)
plt.show()

x = np.array([[4,2],[2,5],[6,4]])
y = - (2 * x) - 12
plt.plot(x,y)
plt.show()
```
